```python
from anypytools import AnyPyProcess, macro_commands
import numpy as np
import seaborn as sns
```

```python

macrolist = []

# Input parameters
MassParameters = np.linspace(0.2, 6, 6)
CONFIGS = ['"Loaded"', '"Unloaded"']

for mass_value in MassParameters:
    for conf in CONFIGS:
        macro = [
            # The CONFIG define statement is set through the load command
            macro_commands.Load("Demo.Arm2D.any", defs={"CONFIG": conf}),
            # The maass value is set through the setvalue command
            macro_commands.SetValue("Main.InputVariables.MassArm", mass_value),
            # Run the Inverse Dynamics analysis
            macro_commands.OperationRun("Main.ArmModelStudy.InverseDynamics"),
            # We have grouped all our input/output in separate folders. 
            # This means we can add new variables without having to change the macro.
            macro_commands.Dump("Main.InputVariables"),
            macro_commands.Dump("Main.ArmModelStudy.Output.OutputVariables"),
            macro_commands.Dump("Main.ArmModelStudy.Output.Abscissa.t"),
        ]
        macrolist.append(macro)
```

```python
app = AnyPyProcess(num_processes=2)
results = app.start_macro(macrolist)
```

```python

# By specifing index_var as the time variables, any variables with a dimension 
# equal to the length of the time variable will split across rows in the dataframe.
# Note: If the time variables are not consiestent across runs, it is possible to interpolate
# the data.
df_all = results.to_dataframe(index_var="Main.ArmModelStudy.Output.Abscissa.t")
df_all.head()
```

```python
# The dataframe contains a lot of extra data (like task meta info) that we don't need.
# So we will clean up the dataframe and rename the columns to make it eaiser to plot. 

df = df_all.copy()
# The index represents the macro run, and the iStep variable. We could rename and use them 
# `df.index.names = ["iRun", "istep"]` but we don't need them here.
df = df.reset_index(drop=True)
df.columns = df.columns.str.replace("Main.InputVariables.", "")
df.columns = df.columns.str.replace("Main.ArmModelStudy.Output.OutputVariables.", "")
df.columns = df.columns.str.replace("Main.ArmModelStudy.Output.Abscissa.t", "time")
df = df.drop(columns=["Main.ArmModelStudy.Output.OutputVariables", "Main.InputVariables"])
df = df.loc[:, ~df.columns.str.startswith('task_')] # Drop all column with meta data
df.head()
```

```python
sns.lineplot(data=df, x="time", y="BicepsShort", hue="Conf", errorbar=lambda x: (x.min(), x.max()))
```

```python
df.drop(columns=["time"]).groupby(["MassArm", "Conf"]).agg(["max", "min"])
```

```python
# We can also "melt" the data to make it event easier to plot
df_melt = df.melt(id_vars=["time", "Conf", "MassArm"], value_vars = ["DeltodeusA", "BicepsShort", "Brachialis"], var_name="Muscle", value_name="Activity")
df_melt.head()
```

```python
fg = sns.relplot(data=df_melt, x="time", y="Activity", col="Muscle", row= "Conf", hue="Muscle", kind="line", height=3)
```
